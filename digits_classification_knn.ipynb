```python
import pandas as pd
import matplotlib.pyplot as plt

#Importing our dataset
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(2):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)
len(X_train)
```

```python
len(X_test)
```

```python
#Using k=3
from sklearn.neighbors import KNeighborsClassifier

model_k3 = KNeighborsClassifier(n_neighbors=3)
model_k3.fit(X_train, y_train)
```

```python
#Using k=4

model_k4 = KNeighborsClassifier(n_neighbors=4)
model_k4.fit(X_train, y_train)
```

```python
#Using k=6

model_k6 = KNeighborsClassifier(n_neighbors=6)
model_k6.fit(X_train, y_train)
```

```python
#Checking the accuracy of our model with k=3

model_k3.score(X_test, y_test)
```

```python
#Checking the accuracy of our model with k=4

model_k4.score(X_test, y_test)
```

```python
#Checking the accuracy of our model with k=6

model_k6.score(X_test, y_test)
```

```python
digits.target[25] #Ground truth for these test values
```

```python
model_k3.predict([digits.data[25]]) #Test prediction with k=3
```

```python
model_k4.predict([digits.data[25]]) #Test prediction with k=4
```

```python
model_k6.predict([digits.data[25]]) #Test prediction with k=6
```

```python
#Using k=3

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_prediction = model_k3.predict(X_test)

accuracy = accuracy_score(y_test, y_prediction)
precision = precision_score(y_test, y_prediction, pos_label=1, average='weighted')
recall = recall_score(y_test, y_prediction, pos_label=1, average='weighted') 
f_score = f1_score(y_test, y_prediction, pos_label=1, average='weighted')

print("For K-Nearest Neighbour (KNN) using k=3")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F-score: {f_score}")
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_prediction)
print('RBF Kernel')
plt.figure(figsize = (6,3))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Prediction")
plt.ylabel("Truth")
```

```python
#Using k=4

y_prediction = model_k4.predict(X_test)

accuracy = accuracy_score(y_test, y_prediction)
precision = precision_score(y_test, y_prediction, pos_label=1, average='weighted')
recall = recall_score(y_test, y_prediction, pos_label=1, average='weighted') 
f_score = f1_score(y_test, y_prediction, pos_label=1, average='weighted')

print("For K-Nearest Neighbour (KNN) using k=4")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F-score: {f_score}")
```

```python
cm = confusion_matrix(y_test, y_prediction)
print('Linear Kernel')
plt.figure(figsize = (6,3))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Prediction")
plt.ylabel("Truth")
```
